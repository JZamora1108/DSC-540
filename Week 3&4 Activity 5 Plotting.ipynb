```python
import numpy as np
import pandas as pd
import requests
import matplotlib.pyplot as plt
import os

os.getcwd()
```

```python
## 1 & 2 import libraries and read csv file

df=pd.read_csv("Boston_housing.csv")

print(df)
```

```python
## 3 View first 10 records

df.head(10)
```

```python
## 3 continued find total records
df.shape
```

```python
## 4 create smaller data fram trimpping chas, nox, b and lstat

df=df.drop(['CHAS', 'NOX', 'B', 'LSTAT'], axis=1)
```

```python
## 5 cheack last 7 records
df.tail(7)
```

```python
## 6 histogram of all columns

df.hist()
plt.show()
```

```python
## 7 plotting set up

column_title = df.columns.array
column_title
```

```python
for col in column_title:
    df[col].hist()
    plt.title("histogram featuring "+ col,fontsize=15)
    plt.show()
```

```python
## 8 scatter plot creation and setup

df.plot.scatter('CRIM', 'PRICE', s=200, c='black', edgecolor='b')
plt.grid(True)
plt.title('Crime Rate Versus Price')
plt.xlabel('Crime Rate')
plt.ylabel('Price')
plt.show()
```

```python
## 9 plot using log10(crime) vs price

plt.scatter(np.log10(df['CRIM']),df['PRICE'], s=200, c='black', edgecolor='r')
plt.grid(True)
plt.title('Log Crime Rate Versus Price')
plt.xlabel('Log Crime Rate')
plt.ylabel('Price')
plt.show()
```
