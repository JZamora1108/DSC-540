```python
##Load data and view 
na_values = [""," ",None,"Missing","NA"]
df=pd.read_csv("candyhierarchy2017.csv",encoding='latin-1',na_values=na_values)
df.head()
```

```python
##Setup data for group by Dicts/Series and cross tabs to fulfill assignment requirements
df.reset_index(inplace=True)
df.head()
```

```python
##View shape
df.shape
```

```python
##Clean by State/Province/City fields that have missing data
df.dropna(subset=['Q4: COUNTRY','Q5: STATE, PROVINCE, COUNTY, ETC'],inplace=True)
df.shape
```

```python
mapping = {
    'Q4: COUNTRY': 'COUNTRY',
    'Q5: STATE, PROVINCE, COUNTY, ETC': 'LOCATION'
}

by_column = df.groupby(mapping, axis=1)
by_column
```

```python
##Breaking down data 
by_column.describe()
```

```python
##remove columns to view specific info and cross tab
##Insert Candy Column
candy_names = {x:x.split('Q6 | ')[1] for x in df.columns if x.startswith('Q6 | ')}
new_index = pd.Index(list(candy_names.values()),name='Candy')
new_df = df.reindex(columns=new_index)
new_df.head()
ldata = new_df.stack().reset_index().rename(columns={0: 'Candy_Approval'})
ldata[:10]
pd.crosstab([ldata['Candy']],ldata.Candy_Approval,margins=True)
```

```python
crosstab_df = pd.crosstab([ldata['candy']],ldata.candy_liking)
print(crosstab_df[crosstab_df['JOY'] == crosstab_df['JOY'].max()])
```
